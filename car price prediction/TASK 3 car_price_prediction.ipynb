```python
labels=car["price"].to_numpy()
car= car.drop(columns=["car_ID", "CarName", "doornumber", "price"])
```

```python
def replace_with_num(column, legend):
    car[column]=[legend[val] for val in car[column]]

mappings={"gas": 0, "diesel": 1,
            "std": 0, "turbo": 1,
            "front": 0, "rear": 1,
            "rwd": 0, "fwd": 1, "4wd": 2}

replace_with_num("fueltype", mappings)
replace_with_num("aspiration", mappings)
replace_with_num("enginelocation", mappings)
replace_with_num("drivewheel", mappings)

carbody_map={carbody: i for i, carbody in enumerate(car["carbody"])}
replace_with_num("carbody", carbody_map)

fuelsystem_map = {fuelsystem: i for i, fuelsystem in enumerate(car["fuelsystem"])}
replace_with_num("fuelsystem", fuelsystem_map)

cylindernumber_map = {'four': 4, 'six': 6, 'five': 5, 'three': 3, 'twelve': 12, 'two': 2, 'eight': 8}
replace_with_num("cylindernumber", cylindernumber_map)

enginetype_map = {enginetype: i for i, enginetype in enumerate(car["enginetype"])}
replace_with_num("enginetype", enginetype_map)
```

```python
# turn dataframe to matrix and insert x0 as first feature.
input_matrix = car.to_numpy()
input_matrix = np.insert(input_matrix, 0, 1, axis=1)
print("input_matrix shape:", input_matrix.shape)
input_matrix
```

```python
class LinearRegression:
    def __init__(self, X, labels, parameters, learning_rate = 1e-10):
        self.X = X
        self.labels = labels
        self.parameters = parameters
        self.learning_rate = learning_rate

    def predict(self, features):
        return np.dot(self.parameters, features)

    def cost(self):
        cost = 0
        for xi, yi in zip(self.X, self.labels):
            cost += (self.predict(xi) - yi) ** 2
        return cost * 0.5

    def batch_descent(self, epoch = 100_000):
        for _ in range(epoch):
            gradient = np.zeros(self.parameters.shape)
            for xi, yi in zip(self.X, self.labels):
                gradient += (self.predict(xi) - yi) * xi
            self.parameters -= gradient * self.learning_rate
```

From the above code, we can learn the implementation of a simple linear regression algorithm using batch gradient descent.

Here's a breakdown of the code and what we can learn from each part:

1. Class Definition:
   - The code defines a class named `LinearRegression`.
   - The class has an initializer (`__init__` method) that takes the following parameters:
     - `X`: Input features or independent variables.
     - `labels`: Target variable or dependent variable.
     - `parameters`: Coefficients or weights of the linear regression model.
     - `learning_rate`: Learning rate for gradient descent (default value is 1e-10).

2. Prediction:
   - The class has a `predict` method that takes the input features (`features`) and performs the prediction by calculating the dot product of the parameters and features.

3. Cost Calculation:
   - The class has a `cost` method that calculates the cost function, which measures the error or discrepancy between the predicted values and the actual labels.
   - It iterates over the input features (`X`) and corresponding labels (`labels`) to calculate the sum of squared errors.

4. Batch Gradient Descent:
   - The class has a `batch_descent` method that performs batch gradient descent to update the parameters and minimize the cost function.
   - It takes an optional parameter `epoch` that specifies the number of iterations for gradient descent (default value is 100,000).
   - In each iteration, it calculates the gradient of the cost function with respect to the parameters and updates the parameters by subtracting the gradient multiplied by the learning rate.
   - It iterates over the input features (`X`) and corresponding labels (`labels`) to calculate the gradient for each data point and updates the parameters accordingly.

Overall, my code demonstrates the implementation of a basic linear regression model using **batch gradient descent** to find the optimal parameters that minimize the cost function. It shows the key steps involved in training the model and making predictions.

```python
params = np.zeros(23)
model = LinearRegression(input_matrix, labels, params)
model.batch_descent()
print(f"params={model.parameters}")
```

In my above code snippet, a `LinearRegression` model object is created, initialized with a zero-filled parameter vector (`params`) of length 23. Then, the `batch_descent` method is called to train the model using batch gradient descent. Finally, the trained model parameters are printed.

Here's a breakdown of the code:

1. `params = np.zeros(23)`: A numpy array `params` is created with 23 elements, all initialized to zero. This represents the initial parameter vector for the linear regression model.

2. `model = LinearRegression(input_matrix, labels, params)`: An instance of the `LinearRegression` class is created, passing the `input_matrix`, `labels`, and `params` as arguments. The model object is initialized with these values.

3. `model.batch_descent()`: The `batch_descent` method of the `model` object is called to perform batch gradient descent and update the model parameters based on the training data (`input_matrix` and `labels`).

4. `print(f"params={model.parameters}")`: After the training is completed, the final model parameters are printed. The `model.parameters` attribute contains the updated parameter vector obtained after running the batch gradient descent algorithm.

By calling `model.batch_descent()`, the linear regression model is trained to find the optimal parameter values that minimize the cost function (defined in the `cost` method of the `LinearRegression` class). The updated parameters are then printed to observe the learned coefficients.

```python
def compare_prediction(xi, yi):
    prediction = model.predict(xi)
    print(f"prediction: ${prediction:.0f} actual: ${yi:.0f} error: ${abs(prediction-yi):.0f}")
```

The `compare_prediction` function in the given code snippet compares the predicted value of a data point (`xi`) with its actual label (`yi`). It then prints the prediction, actual value, and the absolute error between the prediction and the actual value.

Here's a breakdown of the function:

1. `def compare_prediction(xi, yi):`: This line defines a function named `compare_prediction` that takes two arguments: `xi` (a data point) and `yi` (the corresponding actual label).

2. `prediction = model.predict(xi)`: It uses the `model` object to predict the value for the data point `xi` using the `predict` method. The predicted value is stored in the variable `prediction`.

3. `print(f"prediction: ${prediction:.0f} actual: ${yi:.0f} error: ${abs(prediction-yi):.0f}")`: It prints the prediction, actual value, and the absolute error between them. The `f-string` format is used to format the output with appropriate placeholders and precision.

   - `${prediction:.0f}` formats the predicted value (`prediction`) as a floating-point number with no decimal places.
   - `${yi:.0f}` formats the actual value (`yi`) as a floating-point number with no decimal places.
   - `${abs(prediction-yi):.0f}` formats the absolute error between the predicted value and actual value as a floating-point number with no decimal places.

The purpose of this function is to compare individual predictions with their actual values and print the prediction, actual value, and error for analysis and evaluation. It can be used to assess the performance of the model on specific data points and examine the discrepancy between predictions and actual values.

In my above code snippet, it calculates the average accuracy of the model based on the predicted values compared to the actual labels. Here's how it works:

1. `n = input_matrix.shape[0]`: This line calculates the number of data points in the dataset, which is equivalent to the number of rows in the `input_matrix` (assuming each row represents a data point).

2. `avg_discrepancy = 0`: Initializes the variable `avg_discrepancy` to 0. This variable will store the cumulative discrepancy between the predicted values and the actual labels.

3. `for xi, yi in zip(input_matrix, labels):`: This loop iterates over each data point in the `input_matrix` and its corresponding label in `labels`.

4. `avg_discrepancy += (model.predict(xi) - yi) / yi * 100`: For each data point, it calculates the discrepancy between the predicted value (`model.predict(xi)`) and the actual label (`yi`). The discrepancy is then divided by the actual label (`yi`) and multiplied by 100 to get the percentage difference. The calculated discrepancy is added to the `avg_discrepancy` variable.

5. `avg_discrepancy /= n`: After iterating over all the data points, the `avg_discrepancy` is divided by the total number of data points (`n`) to calculate the average discrepancy.

6. `print(f"average accuracy: {100 - avg_discrepancy:0.1f}%")`: Finally, it calculates the average accuracy by subtracting the average discrepancy from 100 and prints it as a percentage. This value represents the average accuracy of the model's predictions, with a higher percentage indicating better accuracy.

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import plotly.express as px
import plotly.graph_objects as go
```

```python
X_train, X_test, y_train, y_test = train_test_split(input_matrix, labels, test_size=0.2, random_state=42)
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X_train, y_train)
linear_regression_predictions = linear_regression.predict(X_test)
print(linear_regression_predictions)
linear_regression_accuracy = linear_regression.score(X_test, y_test)
print("Linear Regression Accuracy:", linear_regression_accuracy)
```

```python
random_forest = RandomForestRegressor()
random_forest.fit(X_train, y_train)
random_forest_predictions = random_forest.predict(X_test)
print(random_forest_predictions )
random_forest_accuracy = random_forest.score(X_test, y_test)
print("Random Forest Regression Accuracy:", random_forest_accuracy)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

# For Linear Regression
linear_regression_mse = mean_squared_error(y_test, linear_regression_predictions)
linear_regression_mae = mean_absolute_error(y_test, linear_regression_predictions)

print("Linear Regression MSE:", linear_regression_mse)
print("Linear Regression MAE:", linear_regression_mae)

# For Random Forest Regression
random_forest_mse = mean_squared_error(y_test, random_forest_predictions)
random_forest_mae = mean_absolute_error(y_test, random_forest_predictions)

print("Random Forest Regression MSE:", random_forest_mse)
print("Random Forest Regression MAE:", random_forest_mae)
```
